Step 1 - Load dataset

```python
import pandas as pd

df = pd.read_csv('Combined_dataset.csv')
print("Initial Shape:", df.shape)
```

Step 2 - Explore Data

```python
# View the first 5 rows
print("--- FIRST 5 ROWS ---")
print(df.head())

# View the last 5 rows
print("\n--- LAST 5 ROWS ---")
print(df.tail())

print("\n--- DATASET SHAPE ---")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

print("\n--- COLUMNS ---")
print(df.columns.tolist())

print("\n--- DATA TYPES & INFO ---")
print(df.info())
```

Step 3 - Select columns FIRST to avoid dropping rows due to unrelated missing data

```python
cols_to_keep = ['product_id', 'title', 'final_price', 'rating']
df_subset = df[cols_to_keep]
```

Step 4 - Clean missing values on the subset

```python
df_cleaned = df_subset.dropna()
```

Step 5 - Drop duplicates

```python
df_cleaned = df_cleaned.drop_duplicates()
```

Step 6 - Clean final_price: handle symbols, quotes, and commas

```python
df_cleaned['final_price'] = df_cleaned['final_price'].astype(str).str.replace(',', '', regex=False)
df_cleaned['final_price'] = df_cleaned['final_price'].str.replace('₹', '', regex=False)
df_cleaned['final_price'] = df_cleaned['final_price'].str.replace('"', '', regex=False)
```

Step 7 - Convert to numeric (decimal numbers)

```python
df_cleaned['final_price'] = pd.to_numeric(df_cleaned['final_price'], errors='coerce')
df_cleaned = df_cleaned.dropna(subset=['final_price'])
```

Step 8 - Add dummy quantity and calculate total_amount

```python
df_cleaned['quantity'] = 2
df_cleaned['total_amount'] = df_cleaned['final_price'] * df_cleaned['quantity']
print("\nFinal Shape:", df_cleaned.shape)
```

Step 9 - We use .head() here just to preview the final structure safely

```python
display(df_cleaned.head())
```

Step 10 - Save the entire 1000-row dataset to a new CSV

```python
df_cleaned.to_csv('cleaned_shopping_dataset.csv', index=False)
print("\nExport Complete!")
```
